##  고정수익시장 (Renta fija)
Los instrumentos de renta fija son activos financieros de bajo riesgo que ofrecen rendimientos estables y no implican pérdida del capital, pagando intereses de forma periódica o al vencimiento.
- 위험 낮음  - 안정적인 수익  - 원금 손실 없음  
예시 : Bonos del gobierno (Cetes) 국채, Papel comercial 상업이음,  Pagarés a mediano plazo 중기어음, 
- Pagarés financieros 금융어음 Obligaciones 채권, Udibonos, Ajustabonos, Factoraje 팩토링 
- Arrendamiento financiero 금융리스 
##  변동수익시장 (Renta variable)
La renta variable es un conjunto de activos financieros de alto riesgo que pueden generar altos rendimientos, pero también implican la posibilidad de pérdida del capital y no tienen vencimiento.
- 위험 높음, 높은 수익 가능 ,원금 손실 가능  ,만기 없음

# 금본위제 (Patrón Oro)
El patrón oro es un sistema monetario en el que el valor de la moneda está respaldado por el oro.
- 시작: 1870년경 - 기준: 금
-caracteristica:El valor de la moneda depende del oro ,Tipo de cambio fijo ,Convertibilidad en oro  
 문제점
- Favorece a los países con oro   금 보유 국가 유리, Genera desequilibrios entre países  국가 간 불균형
- Limita la oferta monetaria  화폐 공급 제한, Provoca deflación  디플레이션 발생
# 금환본위제 (1922)El patrón cambio oro
El patrón cambio oro es un sistema monetario en el que las monedas se basan en una divisa fuerte (como la libra esterlina), manteniendo el oro como referencia.- 기준: 파운드, 금은 참고 기준
붕괴 이유
- Pérdida de poder de Inglaterra   영국 영향력 감소, Aparición de nuevos competidores  다른 국가 경쟁
- Cambios en el comercio internacional 무역 구조 변화 ,Pérdida de confianza en la libra  파운드 신뢰 하락
# 달러 본위제 (브레튼우즈) El sistema de patron cambios dolares
:El sistema Bretton Woods es un sistema monetario internacional creado en 1944, basado en el dólar como moneda de referencia, con tipo de cambio fijo y convertibilidad en oro.
특징
- Basado en el dólar (달러 중심),Tipo de cambio fijo (고정환율)
- Convertibilidad en oro (금으로 교환 가능),Supervisado por el FMI  
붕괴 이유
El sistema colapsó debido al exceso de dólares, la inflación, los cambios en el comercio internacional, el aumento en el precio del oro, la inestabilidad cambiaria y los costos de las guerras.
- 무역 구조 변화, 달러 과잉 → 인플레이션, 자본 이동 증가, 금 가격 상승, 환율 불안정, 전쟁 비용 증가
 
El patrón oro y el sistema Bretton Woods son sistemas monetarios internacionales basados en el oro; sin embargo, el patrón oro establecía que el valor de la moneda dependía directamente del oro, mientras que Bretton Woods utilizaba el dólar como moneda de referencia con convertibilidad en oro.

Además, el patrón oro tenía un sistema más rígido con oferta monetaria limitada, mientras que Bretton Woods permitía mayor flexibilidad, aunque mantenía tipos de cambio fijos.

Finalmente, ambos sistemas colapsaron debido a desequilibrios económicos, pero Bretton Woods colapsó principalmente por el exceso de dólares, la inflación y los cambios en el comercio internacional.

# 💰 특별인출권 (DEG: Derechos Especiales de Giro)
Los Derechos Especiales de Giro (DEG) son un activo de reserva internacional creado por el Fondo Monetario Internacional (FMI) en 1968 con el objetivo de resolver problemas de liquidez internacional.
Inicialmente, el DEG estaba vinculado al sistema dólar-oro,  
pero a partir de 1976 pasó a estar basado en una **cesta de monedas**.
(국제 유동성 문제를 해결하기 위해 국제통화기금(IMF)은  
1968년에 **특별인출권(DEG)**이라는 인공적인 통화를 창설하였다.
처음에는 달러-금 기준에 연동되었으나,  
1976년부터는 여러 통화로 구성된 **통화 바스켓(통화 묶음)** 형태로 변경되었다.
DEG는 하나의 **계산 단위(unit of account)**이며,  
환율은 변동적이고 자유롭게 변동한다.)
 특징
- Es una moneda artificial (no física) 인공적으로 생성된 통화 → 필요한 만큼 발행 가능
- Es emitida por el FMI IMF라는 독립 기관에 의해 관리됨
- Funciona como unidad de cuenta 여러 통화로 구성 → 안정성 유지
- Tiene tipo de cambio variable (flotación libre)
- Su valor se basa en una cesta de monedas
 구성 통화 선정 기준 Criterio de selección de monedas
Las monedas que componen la cesta son aquellas que:
- Tienen una participación significativa en el comercio internacional 국제 무역에서 중요한 비중을 차지하는 국가
- Poseen más del 1% del volumen de exportaciones 수출 비중이 1% 이상

:El Sistema Financiero Mexicano es el conjunto de instituciones, mercados e instrumentos que permiten canalizar el ahorro hacia la inversión productiva, con el objetivo de fomentar el crecimiento económico.

## 최고 규제 기관(¿Cuáles son las principales instituciones reguladoras?)
Las principales instituciones reguladoras del Sistema Financiero Mexicano son:
- El Banco de México  멕시코 중앙은행 
- La Secretaría de Hacienda y Crédito Público (SHCP) 재무부

## 🏦 멕시코 중앙은행 (Banco de México)
기능: El Banco de México es una institución autónoma cuya función principal es mantener la estabilidad del poder adquisitivo de la moneda nacional.  
Además, regula el tipo de cambio, actúa como banco de bancos y como prestamista de última instancia.

## 🏛️ 재무부 (SHCP)
La SHCP supervisa, controla y vigila a las instituciones financieras a través de tres comisiones:
금융기관을 감독, 통제 및 감시하며  다음 3개 위원회를 통해 수행:
- La Comisión Nacional Bancaria y de Valores (CNBV) : es la institución encargada de regular a los intermediarios bancarios, financieros no bancarios y el mercado bursátil.
- La Comisión Nacional de Seguros y Fianzas (CNSF) supervisa a las instituciones de seguros y de fianzas, garantizando que cumplan con las leyes.
- La Comisión Nacional del Sistema de Ahorro para el Retiro (CONSAR) : regula el sistema de ahorro para el retiro, supervisando a las AFORE y las SIEFORE.
- CONDUSEF: 금융 서비스 이용자 보호 기관
La CONDUSEF es un organismo encargado de proteger y defender a los usuarios de los servicios financieros.

# 💼 금융 자산 (Los Activos Financieros)

## 자산 분류
¿Cómo se clasifican los activos?
Los activos se clasifican en: Activos reales o tangibles, Activos intangibles, Activos financieros  
실물자산 정의 (activos reales) : Los activos reales son bienes físicos que tienen valor debido a su naturaleza o propiedades, como maquinaria, edificios y terrenos.
무형자산 (activos intangibles)
Los activos intangibles son aquellos que no tienen forma física, como la propiedad intelectual, las marcas y el conocimiento.
금융 자산 (Activos financieros)
Los activos financieros son títulos que otorgan derechos sobre activos reales y permiten obtener un ingreso futuro.

 Diferencia entre activos reales e intangibles
La diferencia es que los activos reales tienen forma física, mientras que los intangibles no.

 ¿Qué tipos de activos financieros existen?
Los activos financieros se clasifican en:
- Renta fija  
- Renta variable  
- Mercado de dinero  
- Mercado de capitales  

# 📉 파생상품 (Derivados)
Los derivados son instrumentos financieros cuyo valor depende de otro activo.금융 자산에서 파생된 자산
종류: 옵션 (opciones), 선물 (futuros), 선도 (forwards), 워런트 (warrants), 스왑 (swaps)
 옵션 vs 선물
Diferencia entre futuros y opciones
En los futuros existe la obligación de cumplir el contrato en una fecha futura.
En las opciones existe el derecho, pero no la obligación, de comprar o vender un activo, mediante el pago de una prima.
